```python
import trackpy as tp
```

```python
images = tp.ImageSequence('../trackpy/tests/video/image_sequence/')
```

```python
s = tp.PandasHDFStore('data.h5')
```

We can loop over calls to locate explicitly:

```python
for image in images[:]:
    s.put(tp.locate(image, 11))
```

We can get the data for a given frame:

```python
s.get(2)
```

Or dump all the data, if memory is sufficient:

```python
s.dump()
```

A new batch API eliminates by slew of kwargs in favor of a single optional ``output`` arg. If ``None`` (default), it will just return the results. If you specify an ``output`` class, it will return ``None``. Any class with a ``put`` method will satisfy it.

```python
s2 = tp.PandasHDFStore('data2.h5')
tp.batch(images[:], 11, output=s2)
```

The ``__iter__`` method lets us pass ``s`` directly to ``link_df_iter``.

```python
for linked in tp.link_df_iter(s, 3, neighbor_strategy='KDTree'):
    s.put(linked)
```
